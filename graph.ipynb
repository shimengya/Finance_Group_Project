```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
```

```python
# Load in csv
firm_df = pd.read_csv("consumer_cyclicals_data/firms_summary.csv")
firm_df
```

```python
# select market capital
firm_copy_df = firm_df.copy()
firm_mkt_cap_df = firm_copy_df.loc[:, lambda df: ['Ticker','Market Cap (Billion)']]
# firm_mkt_cap_df.to_csv('firm_mkt_cap.csv', index=False)
firm_mkt_cap_df
```

```python
# figure(figsize=(20, 6), dpi=80)
plt.plot(firm_mkt_cap_df["Ticker"], firm_mkt_cap_df["Market Cap (Billion)"], color='red',marker='o', label = "Reb_Return")
plt.bar(firm_mkt_cap_df["Ticker"], firm_mkt_cap_df["Market Cap (Billion)"], color='skyblue')

# format
plt.title("Market capital of each firm", fontsize=20)
plt.ylabel("Market Cap (Billion)")
plt.xlabel("Tickers of firms")
plt.legend()
plt.show()
```
